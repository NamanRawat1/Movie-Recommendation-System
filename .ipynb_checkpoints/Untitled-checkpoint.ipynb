```python
import pandas as pd
import numpy as np
r_cols=['user_id','movie_id','rating']
ratings=pd.read_csv('u.data.csv',sep='\t',names=r_cols,usecols=range(3))
m_cols=['movie_id','title']
movies=pd.read_csv('u.item.csv',sep='|',names=m_cols,usecols=range(2))
ratings=pd.merge(movies,ratings)
```

```python
ratings.head()
```

```python
movieRatings=ratings.pivot_table(index=['user_id'],columns=['title'],values='rating')
movieRatings.head()
```

```python
AngryMen=movieRatings['12 Angry Men (1957)']
AngryMen.head(10)
```

```python
similarMovies=movieRatings.corrwith(AngryMen)
similarMovies=similarMovies.dropna()
df=pd.DataFrame(similarMovies)
df.head(10)
```

```python
similarMovies.sort_values(ascending=False)
```

```python
movieStats=ratings.groupby('title').agg(('rating':[np.size,np.mean]))
movieStats.head()
```

```python
popularMovies=movieStats['rating']['size']>=100
movieStats[popularMovies].sort([('rating','mean')],ascending:False)[:15]
```
